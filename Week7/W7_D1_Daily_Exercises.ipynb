🌟 Exercise 2 : Scraping robots.txt from Wikipedia
Instructions
Write a Python program to download and display the content of robot.txt for en.wikipedia.org

```python
# Import the necessary module
from urllib.request import urlopen  # urlopen helps us open URLs

# Define the URL for the robots.txt file
robots_url = "https://en.wikipedia.org/robots.txt"

# Open the URL and fetch the robots.txt content
response = urlopen(robots_url)  # This opens the URL

# Read the content of the file
robots_txt = response.read().decode("utf-8")  # Read and decode the content from bytes to a string

# Display the content of robots.txt
print(robots_txt)  # Print the robots.txt content
```

🌟 Exercise 3 : Extracting Headers from Wikipedia’s Main Page
Instructions
Write a Python program to extract and display all the header tags from en.wikipedia.org/wiki/Main_Page.

```python
# Import the necessary modules
from urllib.request import urlopen  # To fetch the webpage
from bs4 import BeautifulSoup  # To parse the HTML content

# Define the URL for Wikipedia's main page
url = "https://en.wikipedia.org/wiki/Main_Page"

# Fetch the HTML content of the page
response = urlopen(url)  # This opens the URL
html_content = response.read()  # Read the content of the page

# Parse the HTML using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')  # 'html.parser' tells BeautifulSoup to parse the HTML

# Extract all header tags (h1, h2, h3, etc.)
header_tags = []  # List to store all headers

# We loop over all header levels (h1 to h6) to find and store them
for i in range(1, 7):
    headers = soup.find_all(f'h{i}')  # Finds all h1, h2, ..., h6 tags
    for header in headers:
        header_tags.append(header.text.strip())  # Add the text of the header to our list

# Display the extracted headers
for idx, header in enumerate(header_tags, 1):
    print(f"Header {idx}: {header}")  # Print each header with its index
```


🌟 Exercise 5 : Analyzing US-CERT Security Alerts
Instructions
Write a Python program get the number of security alerts issued by US-CERT in the current year.
Source: https://www.cisa.gov/news-events/cybersecurity-advisories?f%5B0%5D=advisory_type%3A93

```python
# Step 1: Import necessary libraries
import requests  # We use requests to fetch the webpage content
from bs4 import BeautifulSoup  # BeautifulSoup to parse the HTML
from datetime import datetime  # To get the current year

# Step 2: Define the URL for US-CERT security alerts
url = "https://www.cisa.gov/news-events/cybersecurity-advisories?f%5B0%5D=advisory_type%3A93"

# Step 3: Fetch the HTML content from the page
response = requests.get(url)  # Send a GET request to the URL
html_content = response.text  # Get the HTML content as text

# Step 4: Parse the HTML using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Step 5: Extract the current year
current_year = datetime.now().year  # Get the current year

# Step 6: Find all elements that contain security alert dates
# (This step depends on the structure of the page, so we'll have to inspect the HTML to find the right tags)
alert_dates = []  # Initialize a list to store alert dates
for date_tag in soup.find_all('time'):  # Assuming the dates are stored in <time> tags
    date_text = date_tag.get_text(strip=True)  # Extract the date as text
    if str(current_year) in date_text:  # Check if the current year is in the date
        alert_dates.append(date_text)  # Add the date to our list if it matches the current year

# Step 7: Count the number of security alerts from the current year
number_of_alerts = len(alert_dates)

# Step 8: Display the result
print(f"Number of US-CERT security alerts in {current_year}: {number_of_alerts}")
```
