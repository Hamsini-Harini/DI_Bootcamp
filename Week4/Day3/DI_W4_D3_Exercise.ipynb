#ðŸŒŸ Exercise 6 : Export a dataframe to excel format and JSON format.
Create a simple dataframe.
Export the dataframe to an excel file.
Export the dataframe to a JSON file.

```python
import pandas as pd

# Creating a simple DataFrame
data = {
    'Product': ['Widget A', 'Widget B', 'Gadget C', 'Gizmo D'],
    'Price': [19.99, 24.99, 14.99, 29.99],
    'Quantity': [100, 150, 200, 120],
    'Category': ['Tools', 'Tools', 'Gadgets', 'Gadgets']
}

df = pd.DataFrame(data)
df.head()
```

```python
df.to_excel('products.xlsx', index = False)
df.to_json('products.json', indent = 4)
```

#ðŸŒŸ Exercise 7: Reading JSON Data
Use a sample JSON dataset

Import the JSON data from the provided URL.
Use Pandas to read the JSON data.
Display the first five entries of the data.

Even though there is an option to directly read (load) JSON from URL, the "correct" way to do it via requests library to fetch URL which will give you more possibilites (error handling, processing JSON before loading it, inspecting before loading etc).

Direct Method (pd.read_json() or pd.read_csv()): Use this when the data is clean and can be directly interpreted by Pandas, which is often the case with simple APIs or straightforward CSV/JSON files.

```python

# Directly read the JSON data from the URL into a DataFrame
url = 'https://jsonplaceholder.typicode.com/posts'
df = pd.read_json(url)

# Display the first five entries of the data
df.head()
```

```python
import requests

# Step 1: Import the JSON data from the provided URL
url = 'https://jsonplaceholder.typicode.com/posts'
response = requests.get(url)
json_data = response.json()

# Step 2: Use Pandas to read the JSON data
df = pd.DataFrame(json_data)

# Step 3: Display the first five entries of the data
df.head()
```
