What is Data Analysis?
Data analysis is the process of cleaning, transforming, and interpreting data to uncover useful insights. It helps make sense of large amounts of information, turning it into actionable knowledge. DA can be split in 4 different analysis types:

Descriptive Analysis: Understanding past data.
Predictive Analysis: Forecasting future trends.
Prescriptive Analysis: Recommending actions based on data.

Why is Data Analysis Important?
Data analysis is crucial in todayâ€™s digital world due to:

Informed Decision-Making: It helps businesses base decisions on evidence and trends rather than guesses.
Increased Efficiency: Identifying inefficiencies and optimizing resources through data.
Competitive Advantage: Companies that utilize data stay ahead by understanding markets, customers, and trends faster than competitors.
Areas Where Data Analysis is Applied

Healthcare:
Predict disease outbreaks, personalize treatment plans, and optimize hospital operations.
Example: During COVID-19, data helped track the virus and optimize vaccine distribution.

Finance:
Manage risks, detect fraud, and improve investment strategies.
Example: Banks use algorithms to detect fraud and assess credit risk.

Retail/E-commerce:
Personalize customer experiences, adjust pricing, and manage inventory efficiently.
Example: Amazon analyzes customer behavior to recommend products and optimize stock.

Public sector:
What kinds of laws should we enact - based on what data.
Or another use case would be improving urban living - for example identifying places which need more parking lots etc...

```python
# Mean
mean_sepal_length = db_iris['SepalLengthCm'].mean()
print(f"Mean of Sepal Length: {mean_sepal_length}")

# Median
median_sepal_length = db_iris['SepalLengthCm'].median()
print(f"Median of Sepal Length: {median_sepal_length}")

# Mode)
mode_sepal_length = db_iris['SepalLengthCm'].mode()[0]  # Get the first mode in case of multiple modes
print(f"Mode of Sepal Length: {mode_sepal_length}")
```

```python
import matplotlib.pyplot as plt

# Create a histogram of the 'SepalLengthCm' column
plt.figure(figsize=(8, 6))
plt.hist(db_iris['SepalLengthCm'], bins=10, color='blue', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

The distribution is not normal, it looks skewed to the left, having 2 or 3 peaks, indicating that we might have diffeerent (sub)species in our database
