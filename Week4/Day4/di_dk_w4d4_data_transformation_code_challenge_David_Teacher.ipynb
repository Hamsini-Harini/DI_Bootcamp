```python
# Create a DataFrame from the dataset and start exploration

import pandas as pd

df = pd.read_csv('/content/chiang_mai_1998-2019.csv')

print(df.info())

print(f"This is the shape of the DataFrame : {df.shape}")

df.head()
```

```python
# Any null values?
print(f"There is a total of {df.isnull().sum().sum()} null value(s) in this DataFrame.")
```

## Data Transformation

```python
# Let's perform a Min-Max normalization

from sklearn.preprocessing import MinMaxScaler

#Initiate a MinMaxScaler
scaler = MinMaxScaler()

# Perform Min-Max normalization on the 'Precipitation' column
df['Precipitation_normalized'] = scaler.fit_transform(df[['Precipitation']])
```

```python
# Bonus: Let's see how different are the values by looking at some stats:
print(df[['Precipitation', 'Precipitation_normalized']].describe())
```

##Data Reduction

```python
# Apply PCA to reduce the dimensionality of the dataset

from sklearn.decomposition import PCA

# Initialize PCA to reduce to 2 dimensions
pca = PCA(n_components=2)

# Fit and transform the data
reduce_data = pca.fit_transform(df)
```

```python
# Bonus : Let's got a bit further and concatenate

# We create a DataFrame from the reduced data
pca_df = pd.DataFrame(reduce_data, columns=['PC1', 'PC2'])

# We concatenate our original df with the reduced data
df_with_pca = pd.concat([df.reset_index(drop = True), pca_df], axis = 1)

df_with_pca
```
