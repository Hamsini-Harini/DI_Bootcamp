üåü Exercise 2 : Creating a Line Plot for Temperature Variation
Instructions

Create a simple line plot using Matplotlib that represents the temperature variations over a week.

Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]) and label the x-axis as ‚ÄúDay‚Äù and the y-axis as ‚ÄúTemperature (¬∞F)‚Äù.

```python
import matplotlib.pyplot as plt

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
temperatures = [72, 74, 76, 80, 82, 78, 75]  # Temperatures in ¬∞F


plt.plot(days, temperatures, marker='o', linestyle='-', color='blue', label='Temperature')
# The 'plot' function draws the line graph,
# we use 'marker' to show points at each temperature, 'linestyle' for the line, and 'color' to set the line color.

plt.ylabel('Temperature (¬∞F)')

plt.title('Temperature Variation Over a Week')

plt.legend()  # We display the label defined in the 'plot' function

# Display the final plot
plt.show()
```

üåü Exercise 4 : Data Visualisation

Instructions
Using the dataset sales_data.csv

Calculate the total quantity of products sold by the company.
Identify the category that had the highest revenue and calculate how much revenue it generated.
Determine the average revenue per sale in the dataset.
Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).

```python
import pandas as pd
sales_data = pd.read_csv('sales_data.csv')
sales_data.head() # we could have used as well .info .describe .isnull .duplicated().sum() .unique()
```

```python
total_quantity_sold = sales_data['quantity'].sum()

print(f'Total quantity of products sold: {total_quantity_sold}')
```

```python
# Group by category and calculate total revenue
category_revenue = sales_data.groupby('category')['revenue'].sum()

# Sort the results in descending order
category_revenue = category_revenue.sort_values(ascending=False)

# Get the category with the highest revenue
highest_revenue_category = category_revenue.index[0]
highest_revenue_amount = category_revenue.iloc[0]

print(f"The category with the highest revenue is: {highest_revenue_category} with a total revenue of ${highest_revenue_amount}")
```

```python
# Calculate total revenue
total_revenue = sales_data['revenue'].sum()

# Calculate total quantity sold
total_quantity = sales_data['quantity'].sum()

# Calculate average revenue per sale
average_revenue_per_sale = total_revenue / total_quantity

average_revenue_per_sale
```

```python
# Step 4: Create a bar chart to visualize total revenue by quarter

# Step 4.1: Convert 'date' column to datetime type
sales_data['date'] = pd.to_datetime(sales_data['date'])

# Step 4.2: Extract the month from the date
sales_data['month'] = sales_data['date'].dt.month

# Step 4.3: Create a 'quarter' column to categorize months into quarters
sales_data['quarter'] = pd.cut(sales_data['month'],
                               bins=[0, 3, 6, 9, 12],
                               labels=['Q1', 'Q2', 'Q3', 'Q4'])

# Step 4.4: Group by 'quarter' and sum the revenue for each quarter
quarter_revenue = sales_data.groupby('quarter')['revenue'].sum()

# Step 4.5: Plot the bar chart for quarterly revenue
plt.bar(quarter_revenue.index, quarter_revenue.values, color='lightcoral', edgecolor='black')

# Step 4.6: Label the axes
plt.xlabel('Quarter')
plt.ylabel('Total Revenue ($)')
plt.title('Total Revenue by Quarter')

plt.show()
```

üåü Exercise 5 : Data Visualisation Using MatPlotLib
Instructions
Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10. Add a title and label the x and y axes.
Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week. The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.
Create a pie chart representing the following data about favorite fruits of a group of people: Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%). Customize the chart with different colors for each fruit and add a legend.

```python
import numpy as np  # Importing NumPy to handle the x values

# Step 1: Define the x values ranging from -10 to 10
x: np.ndarray = np.arange(-10, 11, 1)
y: np.ndarray = x ** 2  # The function y = x^2

# Step 2: Create the line plot
plt.plot(x, y, color='purple', linestyle='-', marker='o', label='y = x^2')

# Step 3: Add labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot for y = x^2')


plt.legend()
plt.show()
```

```python
# Step 1: Define the data
products: list[str] = ['A', 'B', 'C', 'D']  # Products
sales: list[int] = [15, 30, 45, 20]  # Sales data

# Step 2: Create the bar chart
plt.bar(products, sales, edgecolor='black')

# Step 3: Add labels and title
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Sales Data for Products A, B, C, and D')

plt.show()
```

```python
# Step 1: Define the data
fruits: list[str] = ['Apples', 'Bananas', 'Cherries', 'Dates']
percentages: list[int] = [40, 30, 20, 10]  # Percentage distribution of favorite fruits
colors: list[str] = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']  # Colors for each fruit

# Step 2: Create the pie chart
plt.pie(percentages, labels=fruits, colors=colors, autopct='%1.1f%%', startangle=140)

plt.title('Favorite Fruits Distribution')
plt.show()
```

Task 4: Regional Comparisons of GDP and Health.

```python

# Step 2: Group the data by 'Country or region' and calculate mean 'GDP per capita' and 'Healthy life expectancy'
# This will allow us to compare these metrics across different countries or regions
grouped_df = df.groupby('Country or region')[['GDP per capita', 'Healthy life expectancy']].mean()

# Step 3: Create a subplot with bar plot for 'GDP per Capita' and line plot for 'Healthy Life Expectancy'
fig, ax1 = plt.subplots(figsize=(14, 8))  # Create a figure and a set of subplots

# Step 4: Bar plot for 'GDP per Capita'
ax1.bar(grouped_df.index, grouped_df['GDP per capita'], color='blue', label='GDP per Capita')
ax1.set_xlabel('Country or Region', fontsize=12)
ax1.set_ylabel('GDP per Capita', fontsize=12)
ax1.set_title('Comparison of GDP per Capita and Healthy Life Expectancy by Country or Region', fontsize=14)
ax1.tick_params(axis='x', rotation=90)  # Rotate x-axis labels for better readability

# Step 5: Create a second y-axis for 'Healthy Life Expectancy'
ax2 = ax1.twinx()  # Create a twin y-axis that shares the same x-axis as ax1
ax2.plot(grouped_df.index, grouped_df['Healthy life expectancy'], color='green', label='Healthy Life Expectancy', marker='o')
ax2.set_ylabel('Healthy Life Expectancy', fontsize=12)

# Step 6: Add grid and layout adjustments
fig.tight_layout()  # Automatically adjust subplot params to give specified padding
ax1.grid(True)

plt.show()
```
