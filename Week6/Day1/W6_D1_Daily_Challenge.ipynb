1. Data Preparation:

Hint 1: Use np.random.uniform(low, high, size) to generate the temperature data.
Hint 2: Create a DataFrame using pd.DataFrame(data, index, columns) with appropriate index and columns.

Use NumPy to generate a synthetic dataset representing average monthly temperatures (in degrees Celsius) for 12 months across 10 different cities. The temperatures should range from -5 to 35 degrees.

Convert this NumPy array into a Pandas DataFrame, adding city names as index and months as columns.


**Steps**:

Use NumPy to generate the synthetic data.
Create a Pandas DataFrame with that data.
Label the index with city names and the columns with month names.

```python
import numpy as np
import pandas as pd

# Step 1: Generate random temperature data using np.random.uniform
# Temperatures will range from -5 to 35 for 12 months across 10 cities
temperature_data = np.random.uniform(low=-5, high=35, size=(10, 12))
# if you print this you get huge array

# Step 2: Create a Pandas DataFrame from the NumPy array
# Defining city names and month names
cities = [f"City {i}" for i in range(1, 11)]  # generate generic cities - City 1 to City 10
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Step 3: Create the DataFrame
temperature_df = pd.DataFrame(temperature_data, index=cities, columns=months)
temperature_df
```

2. Data Analysis:

Hint 1: Calculate the annual average temperature using DataFrame.mean(axis).
Hint 2: Find the city with the highest and lowest average temperature using idxmax() and idxmin() methods.

Calculate the annual average temperature for each city.

Identify the city with the highest and lowest average temperature for the year.

```python
# Step 1: Calculate the annual average temperature for each city
# axis=1 calculates the mean across the columns (i.e., across the months for each city)
annual_avg_temp = temperature_df.mean(axis=1)

# Step 2: Find the city with the highest average temperature
city_with_highest_avg_temp = annual_avg_temp.idxmax()

# Step 3: Find the city with the lowest average temperature
city_with_lowest_avg_temp = annual_avg_temp.idxmin()

# Step 4: Display the results
print("Annual Average Temperature for Each City:")
print(annual_avg_temp)
print("\nCity with the Highest Average Temperature:", city_with_highest_avg_temp)
print("City with the Lowest Average Temperature:", city_with_lowest_avg_temp)
```

```python
# Step 1: Import Matplotlib for visualization
import matplotlib.pyplot as plt

# Step 2: Plot the data
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot a line for each city
for city in temperature_df.index:
    plt.plot(temperature_df.columns, temperature_df.loc[city], label=city)  # Plotting each city's data

# Step 3: Add labels and a title
plt.title("Monthly Temperature Trends for Each City")  # Title for the plot
plt.xlabel("Month")  # X-axis label
plt.ylabel("Temperature (Â°C)")  # Y-axis label

# Step 4: Add a legend to indicate the city names
plt.legend(title="Cities", bbox_to_anchor=(1.05, 1), loc='upper left')

# Step 5: Display the plot
plt.grid(True)  # Add grid for better readability
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
