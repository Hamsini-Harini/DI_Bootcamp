Exercise 1: Linear Regression Analysis (Coding Exercise)
Objective: Implement linear regression on a new dataset and interpret the results.

Instructions:

1. Use the following dataset: x = np.array([2, 4, 6, 8, 10])`, `y = np.array([3, 5, 7, 9, 11]).

2. Apply linear regression using SciPy to find the slope and intercept.

3. Plot the data points and the regression line.

4. Interpret the slope and intercept in the context of this data.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

x = np.array([2, 4, 6, 8, 10])
y = np.array([3, 5, 7, 9, 11])

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
```

```python
# Generate predicted y values from the regression line
y_pred = slope * x + intercept

# Plotting the data points
plt.scatter(x, y, color='blue', label='Data Points')

# Plotting the regression line
plt.plot(x, y_pred, color='red', label=f'Regression Line (y={slope:.2f}x + {intercept:.2f})')

# Adding labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Analysis')

# Show legend and plot
plt.legend()
plt.grid(True)
plt.show()
```

Instructions

Objective: Apply linear regression to a dataset and interpret the results.

Task: Given a dataset of housing prices (house_prices) and their corresponding sizes (house_sizes), use linear regression to predict the price of a house given its size.

Dataset:

house_sizes: [50, 70, 80, 100, 120] (in square meters)
house_prices: [150,000, 200,000, 210,000, 250,000, 280,000] (in currency units)
Questions:

What is the slope and intercept of the regression line?
Predict the price of a house that is 90 square meters.
Interpret the meaning of the slope in the context of housing prices.

```python
import numpy as np  # For numerical operations
from scipy import stats  # For linear regression
import matplotlib.pyplot as plt  # For visualizing

# Given data for house sizes and prices
house_sizes = np.array([50, 70, 80, 100, 120])  # in square meters because Numpy matrices don't have colum names etc like pandas df
house_prices = np.array([150000, 200000, 210000, 250000, 280000])  # in currency units

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

# Plot the data points and regression line
plt.scatter(house_sizes, house_prices, color='blue', label='Data points')  # Scatter plot of original data

# Calculate the predicted prices using the regression line
predicted_prices = slope * house_sizes + intercept

# Plot the regression line
plt.plot(house_sizes, predicted_prices, color='red', label='Regression line')

# Add labels and title
plt.xlabel('House Size (square meters)')
plt.ylabel('House Price (currency units)')
plt.title('House Price vs Size - Linear Regression')

plt.legend()
plt.show()


```

What is the slope and intercept of the regression line?

```python
# The slope and intercept of the regression line
print("Slope:", slope) # measures the "steepness" of the line
print("Intercept:", intercept) # where the line crosses the y-axis (price)
```

ðŸŒŸ Exercise 2 : Understanding ANOVA
Instructions
Objective: Test understanding of ANOVA and its application.

Task: Three different fertilizers are applied to three separate groups of plants to test their effectiveness. The growth in centimeters is recorded.

Dataset:

fertilizer_1: [5, 6, 7, 6, 5]
fertilizer_2: [7, 8, 7, 9, 8]
fertilizer_3: [4, 5, 4, 3, 4]
Questions:

Perform an ANOVA test on the given data. What are the F-value and P-value?
Based on the P-value, do the fertilizers have significantly different effects on plant growth?
Explain what would happen if the P-value were greater than 0.05.

```python
# Growth in cm for each fertilizer group
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# Perform one-way ANOVA
f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

# Print the results
print("F-value:", f_value)
print("P-value:", p_value)
```

Since the p-value (0.00002754) is much smaller than 0.05, we reject the null hypothesis. This means that at least one of the fertilizers has a significantly different effect on plant growth compared to the others.

If the p-value had been greater than 0.05, we wouldn't have enough evidence to say that the fertilizers have different effects because differences in plant growth could just be due to random chance

Instructions
Work with a binomial distribution: calculate probabilities for different numbers of successes.
Example: Calculating the probability of getting exactly 5 heads in 10 coin flips.

```python
from scipy.stats import binom

n = 10  # Total number of trials (coin flips)
p = 0.5  # Probability of success (getting heads in one flip)

# How to calculate probability to get exactly 5 heads (exactly)

k = 5  # Number of successes (heads)
prob_5_heads = binom.pmf(k, n, p)
print("Probability of getting exactly 5 heads:", prob_5_heads)
```

Instructions

Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset.
Example in code: data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr

data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

#the clasic correlation - perason
pearson_corr, _son_p = pearsonr(data['age'], data['income'])
#The pearsonr function returns two values: the correlation coefficient and the p-value
#(which indicates whether the correlation is significant).
print("Pearson Correlation Coefficient:", pearson_corr)

# Calculate Spearman correlation coefficient -one based on rankings
spearman_corr, _ = spearmanr(data['age'], data['income'])
print("Spearman Correlation Coefficient:", spearman_corr)
```
