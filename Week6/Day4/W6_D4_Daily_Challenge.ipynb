```python
import numpy as np
import pandas as pd

# Seed for reproducibility
np.random.seed(0)

# Number of entries
n_entries = 1000

# Simulating dataset
time = np.linspace(0, 100, n_entries)  # Time variable
temperature = 20 + 5 * np.sin(np.pi * time / 50) + np.random.normal(0, 0.5, n_entries)  # Temperature (with some noise)
pressure = 1013 + 20 * np.cos(np.pi * time / 25) + np.random.normal(0, 1, n_entries)  # Pressure (with some noise)
chemical_concentration = 5 + 2 * np.sin(np.pi * time / 10) + np.random.normal(0, 0.2, n_entries)  # Chemical concentration

# Creating DataFrame
ninja_data = pd.DataFrame({
    'Time': time,
    'Temperature': temperature,
    'Pressure': pressure,
    'Chemical Concentration': chemical_concentration
})

ninja_data.head()
```

**Task 1: Data Transformation**

```python
# Normalize the data (mean = 0, std = 1)
ninja_data_normalized = ninja_data.copy()  # Making a copy of the original data
for col in ['Temperature', 'Pressure', 'Chemical Concentration']:
    mean = ninja_data[col].mean()  # Calculating the mean
    std = ninja_data[col].std()  # Calculating the standard deviation
    ninja_data_normalized[col] = (ninja_data[col] - mean) / std  # Applying the formula for normalization

ninja_data_normalized.head()
```

```python
# Applying Fourier transformation on the time-series data
temperature_fft = np.fft.fft(ninja_data['Temperature'])
pressure_fft = np.fft.fft(ninja_data['Pressure'])
chem_concentration_fft = np.fft.fft(ninja_data['Chemical Concentration'])


# Fourier transformed data contains complex numbers, we will take the magnitude
temperature_magnitude = np.abs(temperature_fft)
pressure_magnitude = np.abs(pressure_fft)
chem_concentration_magnitude = np.abs(chem_concentration_fft)
```

 Statistical Analysis:

```python
ninja_data.describe()
```

```python
correlation_matrix = ninja_data.corr()
correlation_matrix
```

```python
# 3. Linear Regression between Temperature and Pressure
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(ninja_data['Temperature'], ninja_data['Pressure'])
print("\nLinear Regression Results:")
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")
print(f"P-value: {p_value}")
```

```python
# R-squared coefficient of determination, measures how well the independent variable (temperature) explains the variation in the dependent variable (pressure).
#in this case it's 0.00005 which is almost nothing
# so we can conclude that temperature does not determine the variation in pressure
#This p-value is much higher than 0.05, which means the relationship between temperature and pressure is not statistically significant.
```

Heatmaps for Correlation Matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Heatmap of Correlation Matrix")
plt.show()
```
